```python
# Problem 2
first = str(input("Enter your first name: ")).lower()
last = str(input("Enter your last name: ")).lower()

if first == "alex" or last == "oliver":
    print("You are my friend!")
else:
    print("Keep trying!")
```

```python
# Problem 3
while True:
    try: 
        num1 = float(input("Enter a float number: "))
        break  # Exit the loop if input is valid
    except ValueError:
        print("Invalid input. Please enter a float number.")

print(f"You entered: {num1}")

```

```python
# Problem 4
try:
    num1 = float(input("Enter a float number: "))
    num1 = abs(num1) # absolute value
    print(f"The absolute value is: {num1}")
except ValueError:
    print("Invalid input. Please enter a float number.")
```

```python
# Problem 5
age = int(input("Enter your age: "))

if age < 18:
    print("No driver license")
elif age >= 18 and age <= 60:
    print("You can apply for drive license")
elif age > 60:
    print("You have lots of experience")
```
