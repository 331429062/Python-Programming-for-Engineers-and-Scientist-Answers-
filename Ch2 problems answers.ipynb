```python
# Problem 2:
num = int(input("Enter you age: "))
if num >= 18:
    print("Yes you can take a driver's license")
elif num < 18 and num > 0:
    print("No driver license")
else:
    print("Invalid input")
# Display the image in the notebook
display(Image(filename=r"e:\Python\Python for Engineers and Scientist\Ch2\flowchart 2.2.png"))
```
